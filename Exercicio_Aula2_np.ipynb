### Exercício 1 — Fácil: Criando e inspecionando matrizes

**Enunciado:** Crie as seguintes matrizes e exiba o `shape`, `ndim` e `size` de cada uma:

a) Uma matriz 3×3 com os valores de 1 a 9 (use `arange` + `reshape`).  
b) Uma matriz de zeros com 4 linhas e 2 colunas.  
c) A matriz identidade de tamanho 3.

> 💡 **Dica:** Para o item (a), crie primeiro o array com `np.arange(1, 10)` e depois use `.reshape(3, 3)`.

a) *Resposta*

````python
matriz = np.arange(1, 10)
print(matriz)

matriz_3x3 = matriz.reshape(3, 3)
print('Matriz 3x3: \n', matriz_3x3)

print(f'shape: {matriz_3x3.shape}, size: {matriz_3x3.size}, ndim: {matriz_3x3.ndim}')

````

b) *Resposta*

````python
matriz_zeros = np.zeros((4,2))
print("Matriz_zeros: \n", matriz_zeros)

print(f'shape: {matriz_zeros.shape}, size: {matriz_zeros.size}, ndim: {matriz_zeros.ndim}')
````

c) *Resposta*

````python
matriz_ident = np.eye(3)
print('Matriz Identidade 3: \n', matriz_ident)
print(f'shape: {matriz_ident.shape}, size: {matriz_ident.size}, ndim: {matriz_ident.ndim}')
````

### Exercício 2 — Fácil: Indexação e fatiamento

**Enunciado:** Dada a matriz de temperaturas diárias de 3 cidades ao longo de 5 dias:

```python
temperaturas = np.array([
    [32, 33, 31, 29, 35],   # Cidade A
    [28, 27, 30, 26, 29],   # Cidade B
    [22, 24, 21, 23, 25]    # Cidade C
])
```

Extraia:
- A temperatura da Cidade B no 3º dia (índice 2).
- Todas as temperaturas da Cidade C.
- As temperaturas de todas as cidades no último dia.
- A submatriz das cidades A e B nos primeiros 3 dias.

> 💡 **Dica:** Para "último dia", use índice `-1` na coluna. Para a submatriz, lembre que o fim do slice é exclusivo.

````python
temperaturas = np.array([
    [32, 33, 31, 29, 35],   # Cidade A
    [28, 27, 30, 26, 29],   # Cidade B
    [22, 24, 21, 23, 25]    # Cidade C
])

# Temperatura na Cidade B no 3º Dia
print('Temperatura do 3º Dia na Cidade B: ', temperaturas[1, 2])

#Todas as Temperaturas da Cidade C
print('Temperaturas da Cidade C: ', temperaturas[-1, :])

#Temperaturas de Todas as Cidades no último Dia
print('Temperaturas do último dia em todas as Cidades: ', temperaturas[:, -1])

#submatriz das Cidadades A e B nos 3 primeiros dias
print('Temperaturas das Cidades A e B do primeiro ao terceiro dia: \n', temperaturas[0:2, 0:3])
````

### Exercício 3 — Médio: Reshape, flatten e estatísticas por eixo

**Enunciado:** Você tem as vendas mensais de uma loja ao longo de um ano, registradas em um array 1D de 12 elementos (Janeiro a Dezembro):

```python
vendas_ano = np.array([45, 52, 48, 60, 55, 70, 80, 78, 65, 58, 50, 90])
```

1. Reorganize esse array em uma matriz `(4, 3)` onde cada linha representa um trimestre e cada coluna representa um mês dentro do trimestre.
2. Calcule o total de vendas por trimestre (soma por linha — `axis=1`).
3. Calcule a média de vendas por mês dentro de cada trimestre (já feito pelo item 2 se dividir por 3).
4. Identifique o trimestre com maior faturamento.
5. Use `flatten()` para recuperar o array original.

> 💡 **Dica:** Para o item 4, use `np.argmax()` no array de totais. Lembre que o índice retornado começa em 0 — some 1 para obter o número do trimestre.

````python
vendas_ano = np.array([45, 52, 48, 60, 55, 70, 80, 78, 65, 58, 50, 90])
````

1. *Resposta*

````python
#Array organizado por matriz 4x3

vendas_trimestre = vendas_ano.reshape(4, 3)
print("Vendas Trimestre: \n", vendas_trimestre)
````

2. *Resposta*

````python
#Total de Vendas por Trimestre

tri1 = vendas_trimestre[0, :].sum()
print('Total de Vendas no 1TRI: ', tri1)
tri2 = vendas_trimestre[1, :].sum()
print('Total de Vendas no 2TRI: ', tri2)
tri3 = vendas_trimestre[2, :].sum()
print('Total de Vendas no 3TRI: ', tri3)
tri4 = vendas_trimestre[3, :].sum()
print('Total de Vendas no 4TRI: ', tri4)
````

3. *Resposta*

````python
#Média de Vendas por trimestre

media_tri1 = vendas_trimestre[0, :].mean().round(2)
print('Média de Vendas no 1TRI: ', media_tri1)
media_tri2 = vendas_trimestre[1, :].mean().round(2)
print('Média de Vendas no 2TRI: ', media_tri2)
media_tri3 = vendas_trimestre[2, :].mean().round(2)
print('Média de Vendas no 3TRI: ', media_tri3)
media_tri4 = vendas_trimestre[3, :].mean().round(2)
print('Media de Vendas no 4TRI: ', media_tri4)
````

5. *Resposta*

````python
vendas_year = vendas_trimestre.flatten()
print("Array Original: ", vendas_year)
````

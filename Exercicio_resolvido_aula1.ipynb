**EXERCICIO 3**

**Enunciado:** Uma padaria registrou suas vendas (em R$) de segunda a sábado: `[450.00, 320.50, 510.75, 489.00, 605.25, 780.50]`. Com esse array:

1. Calcule o total faturado na semana.
2. Calcule a média diária de vendas.
3. Identifique o menor e o maior faturamento.
4. Aplique um desconto de 5% em todos os dias (simule uma promoção onde a padaria deu 5% de cashback). Exiba os valores após o desconto.

**EXERCICIO 5**

**Enunciado:** Duas equipes de vendas registraram seus resultados mensais ao longo de 6 meses:

- **Equipe A:** `[42000, 38500, 51000, 47300, 55800, 49200]`
- **Equipe B:** `[39000, 41200, 48700, 52000, 44600, 53100]`

Resolva as seguintes questões **sem usar laços `for`** (use apenas operações NumPy):

```python
equipe_A = np.array([42000, 38500, 51000, 47300, 55800, 49200])
equipe_B = np.array([39000, 41200, 48700, 52000, 44600, 53100])
```

1. Qual equipe teve o maior faturamento **total** em 6 meses?

```python
total_A = np.sum(equipe_A)
total_B = np.sum(equipe_B)

print(f'O total vendido pela equipe A foi de R$ {total_A:,.2f}')
print(f'O total vendido pela equipe B foi de R$ {total_B:,.2f}')

if total_A > total_B:
    print(f'A Equipe A teve o maior faturamento total do semestre vendendo R$ {total_A:,.2f}')
else:
    print(f'A Equipe B teve o maior faturamento total do semestre vendendo R$ {total_B:,.2f}')
```

2. Calcule a diferença de vendas mês a mês (Equipe A − Equipe B). Em quais meses a Equipe A superou a Equipe B?

```python
meses = np.array(['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho'])
diferenca_vendas = equipe_A - equipe_B

print(diferenca_vendas)

for i in np.arange(0, diferenca_vendas.size):
    if diferenca_vendas[i] > 0:
        print(f'A equipe A vendeu mais no mes de {meses[i]} com adicional de R$ {diferenca_vendas[i]:,.2f} vendidos')
    else:
        print(f'A equipe B vendeu mais no mes de {meses[i]} com adicional de R$ {(diferenca_vendas[i] * -1):,.2f} vendidos')
```

3. Calcule a média de cada equipe e o desvio padrão. Qual equipe é mais consistente (menor desvio)?

```python
media_A = np.mean(equipe_A)
media_B = np.mean(equipe_B)

desvio_A = np.std(equipe_A)
desvio_B = np.std(equipe_B)

print(f'Media A: {media_A:,.2f}, \nDesvio A: {desvio_A:,.2f}, \nMedia B: {media_B:,.2f}, \nDesvio B: {desvio_B:,.2f}\n')

if desvio_A < desvio_B:
    print(f'A Equipe A é mais consistente em suas vendas, com média semestral de R$ {media_A:,.2f} e desvio padrão de R$ {desvio_A:,.2f}')
else:
    print(f'A Equipe B é mais consistente em suas vendas, com média semestral de R$ {media_B:,.2f} e desvio padrão de R$ {desvio_B:,.2f}')
```

4. Crie um array `bonus` onde cada equipe recebe 8% do seu próprio faturamento mensal como bônus. Exiba o total de bônus de cada equipe.

```python
bonus_A = equipe_A * 0.08
bonus_B = equipe_B * 0.08

total_bonus_A = np.sum(bonus_A)
total_bonus_B = np.sum(bonus_B)

print(bonus_A, bonus_B)

print(f'A Equipe A ganhou um total de R$ {total_bonus_A} de bonus neste semestre')
print(f'A Equipe B ganhou um total de R$ {total_bonus_B} de bonus neste semestre')
```
